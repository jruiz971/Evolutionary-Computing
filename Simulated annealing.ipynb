# Simulated Annealing

### _Given a finite set of objects, where each object has a specific weight and value, and a backpack that can only hold a certain weight. Find the subset of objects that can be put into the backpack while maximizing their value._

### Considerations:
- The representation of a solution will be a binary string of size $n$. Where each position $i$, indicates with a $1$ that the object i is in the backpack or with a $0$ that it is not.
- The initial solution is generated by choosing objects randomly as long as the capacity of the knapsack is not exceeded.
- The neighborhood of a solution is defined as follows: For each position of the binary string, its value is checked and changed. If the generated solution is feasible (it does not exceed the capacity of the knapsack), then is part of the neighborhood.
- The selection of a solution within the neighborhood will be done randomly. 
- To vary the temperature, the following will be used $T_{(t+1)} = 0.99T(t)$

### The input to the program will be:
1. The first line will have the number of cities N.
2. The second line will have the maximum number of iterations $I_\max$ of the Taboo Search algorithm.
3. The next $N-1$ lines will indicate the cost of going from one city to another. That is:

 - The first line has the cost of going from city $n_0$ to city $n_1$, to city $n_2$ and up to city $n_ {N-1}$.

 - The second line has the cost of going from city $n_1$ to city $n_2$ and to city $n_{N-1}$.

 - So on and so forth until the last line has the cost of going from city $n_{N-2}$ to city $n_{N-1}$.

### The output of the program will be:
1. Path to be followed by the traveler (permutation).
2. Cost of following the path found.
